## Methods (Behavior)


Functions inside class

lets create a discount method

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def apply_discount(self, percent):
        self.price = self.price * (1 - percent/100)

```

```python
p = Product("Laptop", 100000)
p.apply_discount(10)

print(p.price)
```

## Encapsulation (Control Access )

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

```python
my_bankaccount = BankAccount(10000)
```


You can’t directly change bank balance

## Polymorphism (Same Function, Different Behavior)

Same method name, different behavior

```python
class Payment:
    def pay(self):
        print("Processing payment")

class CreditCard(Payment):
    def pay(self):
        print("Paid using Credit Card")

class UPI(Payment):
    def pay(self):
        print("Paid using UPI")

```

```python
for method in [CreditCard(), UPI()]:
    method.pay()
```

Different payment methods → same action (pay())

```python
class MLModel:
    def __init__(self):
        self.__weights = None   # private

    def train(self, X, y):
        self.__weights = sum(y) / sum(X)

    def predict(self, X):
        return [self.__weights * x for x in X]
```
