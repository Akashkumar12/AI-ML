# Data preprocessing

Preprocessing data is essential before analysis to ensure accuracy, reliability, and meaningful insights. Raw data often contains:

- Missing values

- Incorrect data types

- Inconsistent formatting

### lets check for missing values:

## 🧼 Missing Values

```python
# Detect missing values
sales_df.isnull().sum()      # Count of missing values per column. (df.isnull()Returns a DataFrame of True/False)
```

price column has 8 missing values, which can be addressed in multiple ways:
- Removing it, if they are few and won’t impact the analysis.
- filling it with mean or median value.

```python
# Drop rows with any missing values
sales_df = sales_df.dropna(subset=['Price'])

#recheck null values
sales_df.isnull().sum()
```

```python
sales_df.shape
```

Since there are only a few null values, we removed them. If the number of null values was higher, we would consider filling them with,
- mean ( **df['Price'].fillna(df['Price'].mean())** )
- median ( **df['Price'] = df['Price'].fillna(df['Price'].median())** )

## Data wrangling





Feature engineering from 'Date' and 'Time' columns, sales aggregation, and data type handling.



```python
df.info()
```

```python
#convert date column into date type
df['Date'] = pd.to_datetime(df['Date'])
# Feature Engineering
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year


```

```python
df
```

```python
#lets check sales by Months
plt.figure(figsize=(10,6))
df.groupby('month')['Sales'].sum().plot(kind='bar', color='coral')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
df
```

```python
# Sales Aggregation
df_daily_sales = df.groupby('Date')['Sales'].sum().reset_index()
df_weekly_sales = df.groupby(df['Date'].dt.isocalendar().week)['Sales'].sum().reset_index()
df_monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum().reset_index()
df_category_sales = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
df_category_sales = df.groupby('Branch')['Sales'].sum().reset_index()
```

```python
df_category_sales
```

```python
# Data Type Handling (Ensure correct types)
df['day_of_week'] = df['day_of_week'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)


df_daily_sales['Sales'] = df_daily_sales['Sales'].astype(float)
df_weekly_sales['Sales'] = df_weekly_sales['Sales'].astype(float)
df_monthly_sales['Sales'] = df_monthly_sales['Sales'].astype(float)
df_category_sales['Sales'] = df_category_sales['Sales'].astype(float)
```

```python
display(df.head())
display(df_daily_sales.head())
display(df_weekly_sales.head())
display(df_monthly_sales.head())
display(df_category_sales.head())
```

##DATA ANALYSIS

Analyze sales trends over time using the daily, weekly, and monthly sales data. Investigate sales patterns for different product categories. Explore the relationship between sales and other variables. Investigate the impact of loyalty membership on sales.



```python
import matplotlib.pyplot as plt

# Sales Trends Over Time
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(df_daily_sales['Date'], df_daily_sales['Sales'], color='blue')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')

plt.subplot(1, 3, 2)
plt.plot(df_weekly_sales['week'], df_weekly_sales['Sales'], color='green')
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Total Sales')

plt.subplot(1, 3, 3)
plt.plot(df_monthly_sales['Date'].astype(str), df_monthly_sales['Sales'], color='red')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()


# Sales Patterns by Category
plt.figure(figsize=(8, 6))
plt.pie(df_category_sales['Sales'], labels=df_category_sales['Category'], autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Sales Distribution by Category')
plt.axis('equal')
plt.show()

# Sales vs. Day of Week, Month, Year, Hour
plt.figure(figsize=(16, 8))
plt.subplot(2, 2, 1)
df.groupby('day_of_week')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Day of Week')

plt.subplot(2, 2, 2)
df.groupby('month')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Month')

plt.subplot(2, 2, 3)
df.groupby('year')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Year')


plt.tight_layout()
plt.show()

# Loyalty Member Impact
member_sales = df[df['Loyalty Member'] == 'Yes']['Sales'].mean()
non_member_sales = df[df['Loyalty Member'] == 'No']['Sales'].mean()

print(f"Average sales for members: {member_sales}")
print(f"Average sales for non-members: {non_member_sales}")

plt.figure(figsize=(6, 4))
plt.bar(['Members', 'Non-Members'], [member_sales, non_member_sales], color=['skyblue', 'lightcoral'])
plt.title('Average Sales by Loyalty Membership')
plt.ylabel('Average Sales')
plt.show()
```

# Insights, Key Trends, and Recommendations from Café Sales Analysis

## Key Insights from the Data

### 1. **Sales Patterns**
- The average transaction involves **2 items** (mean quantity = 1.95)
- **75% of items sold** are priced below ₹230, suggesting most sales come from mid-range products
- There are **6 distinct menu items** across 2 categories (Bakery and Beverage)

### 2. **Customer Behavior**
- **Loyalty program participation** is mixed (some "Yes", some "No" in Loyalty Member column)
- **Payment methods** vary between Online, Card, and Cash
- Some customers make **repeat purchases** (same Customer ID appears multiple times)

### 3. **Environmental Factors**
- Average temperature during sales was **32.2°C**
- Temperature ranges from **25.1°C to 39.8°C**, suggesting the data covers different seasons

## Notable Trends

### 1. **Product Performance**
- **Bakery items** (Bagel, Croissant, Blueberry Muffin) appear to have higher prices (₹200-250) than beverages (₹150-180)
- **Blueberry Muffin** is the highest priced item at ₹250

### 2. **Time Patterns**
- Transactions occur at all times of day (from early morning to late night)
- The data spans several months (March to May 2025), allowing for seasonal analysis

### 3. **Branch Operations**
- Data covers **3 branches (A, B, C)** with different sales patterns
- Some branches may perform better at different times or with different products



## Recommendations for the Café

### 1. **Menu Optimization**
- **Feature high-margin items** like Blueberry Muffins more prominently
- Consider **bundling strategies** (since average quantity is ~2 items)
- Analyze if **bakery items sell better** at certain times or temperatures

### 2. **Customer Engagement**
- **Increase loyalty program adoption** (only some customers are members)
- **Promote online payments** if they're more efficient than cash
- Study **repeat customer behavior** to improve retention

### 3. **Operational Improvements**
- **Staff scheduling** could be optimized based on sales time patterns
- **Inventory management** should account for:
  - Higher bakery prices but potentially different demand
  - Temperature effects on beverage vs. bakery sales
- **Branch-specific strategies** may be needed based on performance differences
