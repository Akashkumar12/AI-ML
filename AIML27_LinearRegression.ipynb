```python
#import Required Libraries
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
X = np.array([[1],[2],[3]])
y = np.array([2,4,5])
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.predict(np.array([[4]]))
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_predict=model.predict(X)   #y = 0.66 + (1.5) x
y_predict
```

```python
mse = mean_squared_error(y,y_predict)
mse
```

```python
model.score(X,y)
```

```python
#import all required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/auto-mpg.data",names=["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"],sep="\s+",na_values='?')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
X = df[['cylinders', 'displacement', 'horsepower', 'weight','acceleration']]
y = df['mpg']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
lr  = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)  #NO
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_train,y_train)
```

```python
import joblib
joblib.dump(lr,"mymodel")
joblib.dump(scaler,"scalermodel")
```

```python
import gradio as gr
import joblib
def predict_mpg(cylinders,displacement,horsepower,weight,acceleration):
  input_data = [[cylinders,displacement,horsepower,weight,acceleration]]
  model = joblib.load("/content/mymodel")
  scaler = joblib.load("/content/scalermodel")
  input_scaled = scaler.transform(input_data)
  prediction = model.predict(input_scaled)
  return round(prediction[0],2)
app = gr.Interface(
    fn=predict_mpg,inputs=["number","number","number","number","number"],outputs="number",title="Auto MPG Prediction"
)
app.launch()
```
