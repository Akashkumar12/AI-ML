

**Standard Deviation (std)** tells you how much variation there is in your data.

* Low std → values are close to the mean (less variation).

* High std → values are spread out (more variation, possibly with outliers).

What are 25%, 50% and 75% ?



| Percentile | Name                                   | Meaning                                                           | Interpretation in your dataset (example: `Age`)               |
| ---------- | -------------------------------------- | ----------------------------------------------------------------- | ------------------------------------------------------------- |
| **25%**    | **First Quartile (Q1)**                | 25% of the data values are **below** this number.                 | For `Age`, 25% of customers are younger than **28.75 years**. |
| **50%**    | **Second Quartile (Q2)** or **Median** | 50% of the data are below and 50% are above this number.          | For `Age`, the median age is **33.5 years**.                  |
| **75%**    | **Third Quartile (Q3)**                | 75% of the data values are **below** this number (and 25% above). | For `Age`, 75% of customers are younger than **41 years**.    |

Another important step before moving on to analysis is to ensure that all columns have the appropriate data types

```python
#check datatype of each column
df.dtypes
```

```python
#syntax: df['column_name'] = df['column_name'].astype(new_dtype)
```

```python
cols = ['product_category', 'customer_country', 'traffic_source', 'payment_method']

for col in cols:
    df[col] = df[col].astype('category')
```

```python
df['is_returned'] = df['is_returned'].astype(bool)
```

### order_date column -> date type

```python
#convert date column into date type
df['order_date'] = pd.to_datetime(df['order_date'])
```

### **Feature Engineering**

```python
# Feature Engineering
df['day_of_week'] = df['order_date'].dt.dayofweek
df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year
```

## **Visualization** "Make Data Speak"

### Revenue by Country (Bar Chart)

```python
import matplotlib.pyplot as plt

country_revenue.plot(kind="bar")
plt.title("Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.show()
```

That small 1e6 on the top-left means:

“All values on this axis are multiplied by 10
6
 (1,000,000)”

So when you see:

* 8 → it actually means 8,000,000
* 8.5 → 8,500,000

Why Matplotlib does this

* Your revenue values are very large (millions)

* To avoid long labels like 8000000, it uses scientific notation
